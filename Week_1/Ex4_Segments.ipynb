# 📚 Segments

Your task is to **divide a string into segments** so that each segment is a **repeat of the same character**. The result should be represented as a list of **pairs**, where each pair contains:

- the **length** of the segment, and  
- the **character** that is repeated.

---

## 📌 Example

For the input string:

```
aaabbccdddd
```

The output should be:

```python
[(3, 'a'), (2, 'b'), (2, 'c'), (4, 'd')]
```

---

## 🔧 Instructions

1. Create a file named `segments.py`.
2. Implement a function called `find_segments` with the following signature:

```python
def find_segments(s: str) -> list[tuple[int, str]]:
```

3. The function should return a list of tuples (length, character) representing segments of repeated characters in the string.

---

## 🔍 Input Constraints

- Input is a string containing characters `a–z`
- Length of the string: **1 to 100 characters**

---

## ✅ Example Usage

```python
from segments import find_segments

print(find_segments("aaabbccdddd"))
# Output: [(3, 'a'), (2, 'b'), (2, 'c'), (4, 'd')]
```

---

## 🧪 Testing

Your function will be tested using many different strings meeting the above constraints.


````python
def find_segments(data):
    if not data:
        return []
    current_char = data[0]
    count = 1
    segment = []
    for i in range(1, len(data)):
        if data[i] == current_char:
            count+=1
        else:
            segment.append((count,current_char))
            current_char = data[i]
            count = 1
    return segment
        


if __name__ == "__main__":
    print(find_segments("aaabbccdddd"))
    # [(3, 'a'), (2, 'b'), (2, 'c'), (4, 'd')]

    print(find_segments("aaaaaaaaaaaaaaaaaaaa"))
    # [(20, 'a')]

    print(find_segments("abcabc"))
    # [(1, 'a'), (1, 'b'), (1, 'c'), (1, 'a'), (1, 'b'), (1, 'c')]

    print(find_segments("kissa"))
    # [(1, 'k'), (1, 'i'), (2, 's'), (1, 'a')]
````
